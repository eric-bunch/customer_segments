We can see that `Detergents_Paper` and `Grocery` are highly correlated. Before we remove one of these features, we will perform a log transform of the data, and then check again if the transformed data is correlated.

Perform log transform in an attempt to have distributions closer to a normal distribution.

```python
sns.pairplot(
    data=np.log(customers[continuous_features]),
    size=2,
    plot_kws={'s': 8},
    diag_kind='kde',
    vars=continuous_features
);
```

We can see now that in the log transformed data, none of the features are highly correlated, so we will not remove any of them.

```python
log_data = np.log(customers[continuous_features])
```

### Feature relevance

For each continuous feature, we leave it out and fit a decision tree regressor on the remaining features. If the `R^2` value is high, we might leave that feature out since it is predictable from the remaining features.

```python
for feat in continuous_features:
    # make a copy of the DataFrame, using the 'drop' function to drop the given feature
    new_data = log_data[continuous_features].copy().drop(feat, axis=1)
    label = log_data.loc[:,feat]

    # split the data into training and testing sets using the given feature as the target
    X_train, X_test, y_train, y_test = train_test_split(new_data, label, 
                                                        test_size=.25, random_state=168)

    # create a decision tree regressor and fit it to the training set
    regressor = DecisionTreeRegressor(random_state=168);
    regressor.fit(X_train, y_train);

    # report the score of the prediction using the testing set
    score = regressor.score(X_test, y_test)
    print("{FEAT} removed, score: {SCORE}".format(
        FEAT=feat,
        SCORE=score
    ))
```

None of these `R^2` values are high enough to leave out a feature.

## Remove outliers

We remove points that lie outside of the interval [Q1 - 1.5\*IQ, Q3 + 1.5\*IQ] for each feature, where Q1 and Q3 are the first and third percentiles respectively, and IQ is the interquartile distance; i.e. Q3 - Q1.

```python
# Make a dict to store outlier indices for all features
out_ind = {}

# For each feature find the data points with extreme high or low values
for feature in log_data.keys():
    
    # Calculate Q1 (25th percentile of the data) for the given feature
    Q1 = np.percentile(log_data[feature], 25)
    
    # Calculate Q3 (75th percentile of the data) for the given feature
    Q3 = np.percentile(log_data[feature], 75)
    
    # Use the interquartile range to calculate an outlier step (1.5 times the interquartile range)
    step = 1.5 * (Q3 - Q1)
    
    # Display the outliers
    print("Data points considered outliers for the feature '{}':".format(feature))
    log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))]
    
    # Convert index to list
    bool_series = ~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))
    lst = bool_series[bool_series].index.tolist()
    # Store the indices
    out_ind[feature] = lst 
    
# OPTIONAL: Select the indices for data points you wish to remove

# Combine lists into one
lst_all = [v for k, v in out_ind.items()]
# Find the common elements
outliers = list(set(lst_all[0]).intersection(*lst_all))

# Remove the outliers, if any were specified
good_data = log_data.drop(log_data.index[outliers]).reset_index(drop = True)
```

```python
good_data.head()
good_data.describe()
```

# PCA

```python
pca = PCA(n_components=6)
pca.fit(good_data)
```

```python
evr = pca.explained_variance_ratio_
```

```python
fig, ax = plt.subplots(figsize=(20, 7));

pd.DataFrame(pca.components_).rename(
    columns={
        0: 'Fresh',
        1: 'Mild',
        2: 'Grocery',
        3: 'Frozen',
        4: 'Detergents_Paper',
        5: 'Delicatessen'
    },
    index={
        0: 'Dim 1',
        1: 'Dim 2',
        2: 'Dim 3',
        3: 'Dim 4',
        4: 'Dim 5',
        5: 'Dim 6'
    }
).plot(
    kind='bar',
#     figsize=(20, 7),
    ax=ax,
    rot=0
);

# for rect in ax.patches:
#     height = rect.get_height()
#     print(rect.get_x(), height)
    
#     ax.text(rect.get_x(), height, 'label')

for k in range(0, 6):
    ax.text(k - 0.4, 1, 'Explained variance: ' + str(evr[k])[0:5])
```

```python
pd.DataFrame(pca.explained_variance_ratio_.cumsum()).T.rename(
    columns={
        k: 'Dim {DIM}'.format(DIM=k) for k in range(0, 6)
    },
    index={
        0: 'Cumulative variance explained'
    }).T.plot(
    kind='line',
    figsize=(15, 5)
);
```

## Dimension reduction

Fit PCA model with two components, and project data onto those two components.

```python
# Apply PCA by fitting the good data with only two dimensions
pca = PCA(n_components=2)
pca.fit(good_data)

# Transform the good data using the PCA fit above
reduced_data = pca.transform(good_data)

# Create a DataFrame for the reduced data
reduced_data = pd.DataFrame(reduced_data, columns = ['Dimension 1', 'Dimension 2'])
```

```python
reduced_data.head()
```

### Biplot

```python
xvector = pca.components_[0] 
yvector = pca.components_[1]

xs = pca.transform(good_data)[:,0] 
ys = pca.transform(good_data)[:,1]
```

```python
reduced_data.plot(
    kind='scatter',
    x='Dimension 1',
    y='Dimension 2',
    figsize=(14, 10)
);

for i in range(len(xvector)):
    plt.arrow(0, 0, xvector[i]*max(xs), yvector[i]*max(ys), color='r', width=0.0005, head_width=0.1)
    
    plt.text(xvector[i]*max(xs)*1.2, yvector[i]*max(ys)*1.2, list(good_data.columns.values)[i], color='k')
```

## Find clusters on reduced dimensions

```python
def cal_score(n_clstr=None):
    """A function to calculate the silhouette score given the number of clusters"""
    # cluster on dimension reduced data 
    clusterer = KMeans(n_clusters=n_clstr, random_state=0).fit(reduced_data)

    # predict the cluster for each data point
    preds = clusterer.predict(reduced_data)

    # find the cluster centers
    centers = clusterer.cluster_centers_

    # calculate the mean silhouette coefficient for the number of clusters chosen
    score = silhouette_score(reduced_data, preds)
    return score
```

```python
# calculate the score for number of clusters from 1 to 10
# store results in a dict
results = {}
for i in range(2, 11):
    results[i] = cal_score(n_clstr=i)

# convert the dict to a pandas series and sort it
pd.Series(results).sort_values(ascending=False)
```

```python
pd.DataFrame(
    data=list(results.values()),
    index=results.keys(),
    columns=['sil. score']
).plot(
    kind='line',
    figsize=(15, 5)
);
```

The number of clusters with the best silhouette score is 2, so we will choose k=2.

```python
# re-calculate to get the parameters
clusterer = KMeans(n_clusters=2, random_state=0).fit(reduced_data)
preds = clusterer.predict(reduced_data)
centers = clusterer.cluster_centers_
```

```python
reduced_data['cluster_prediction'] = preds
```

## Partition by `Channel`

We will attach the channel feature to the reduced data.

```python
log_data_channel = pd.merge(
    left=log_data,
    right=customers[['Channel']],
    left_index=True,
    right_index=True,
    copy=False
)
```

```python
good_data_channel = pd.merge(
    left=good_data,
    right=log_data_channel,
    on=['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen'],
    how='left',
    copy=False
)
```

```python
reduced_data_channel = pd.merge(
    left=reduced_data,
    right=good_data_channel[['Channel']],
    left_index=True,
    right_index=True,
    copy=False
)
```

```python
reduced_data_channel.head()
```

```python
plt.figure(figsize=(14, 10))

# plot cluster 0
plt.scatter(
    x=reduced_data_channel[
        reduced_data_channel['Channel'] == 1
    ]['Dimension 1'].values,
    y=reduced_data_channel[
        reduced_data_channel['Channel'] == 1
    ]['Dimension 2'].values,
    color='b',
    alpha=0.7
);

# plot cluster 1
plt.scatter(
    x=reduced_data_channel[
        reduced_data_channel['Channel'] == 2
    ]['Dimension 1'].values,
    y=reduced_data_channel[
        reduced_data_channel['Channel'] == 2
    ]['Dimension 2'].values,
    color='r',
    alpha=0.7
);


plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')

plt.title('Channels on PCA reduced data')
```
